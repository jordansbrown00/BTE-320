````python
from random import randint

def quicksort(L):
  if len(L) < 2:
    return L
  else:
    pivot_idx = len(L) // 2 # floor div bc return integer and indexes are int ; #randint(0, len(L)-1) #The position of the pivot item
    pivot = L[pivot_idx] #This is the pivot item


    less, equal, greater = [], [], []

    for item in L:
      if item < pivot:
        less.append(item)
      elif item == pivot:
        equal.append(item)
      else:
        greater.append(item)
    #print(quicksort(less)+ equal + quicksort(greater))
    return quicksort(less)+ equal + quicksort(greater)

quicksort([8,0,2,6,1,7,3])
````

````python
from random import randint

def quicksort(L, reverse=False):
  if len(L) < 2:
    return L
  else:
    pivot_idx = len(L) // 2 # floor div bc return integer and indexes are int ; #randint(0, len(L)-1) #The position of the pivot item
    pivot = L[pivot_idx] #This is the pivot item


    less, equal, greater = [], [], []

    for item in L:
      if item < pivot:
        less.append(item)
      elif item == pivot:
        equal.append(item)
      else:
        greater.append(item)
    if not reverse:
      return quicksort(less)+ equal + quicksort(greater)
    else:
      return quicksort(greater, reverse=True) + equal + quicksort(less, reverse=True)

    #print(quicksort(less)+ equal + quicksort(greater))
    return quicksort(less)+ equal + quicksort(greater)

quicksort([8,0,2,6,1,7,3])
````

````python
class Loan:
  def __init__(self, name: str, ir: float, amount: float, maturity: int) -> None:
    self.name = name
    self.ir = ir
    self.amount = amount
    self.maturity = maturity
    self.org = "N/A"

  def amountIncrease(self, n):
    return self.amount * (1 + self.ir/100) ** n

  def increaseDifference(self, other, n):
    return self.amountIncrease(n) - other.amountIncrease(n)


mortgage = Loan('mortgage', 3.5, 100000.0, 30)
refinanced_mortgage = Loan('refinanced mortgage', 2, 100000.0, 15 )
student_loan = Loan('student loan', 4, 10000.0, 20)
````

````python
class ATM:
  def __init__(self, num, amount):
    self.account_number = num
    self.account_amount = amount

  def deposit(self, amount):
    self.account_amount += amount
    print(f'Current funds: ${self.account_amount}')

  def withdraw(self, amount):
    if amount > self.account_amount:
      print(f'Amount ${amount} exceeds current funds: ${self.account_amount}')
    else:
      self.account_amount -= amount
      print(f'Current balance ${self.account_amount}')

account = ATM(123456789, 1000.0)
account.account_amount = 2000
account.withdraw(1500)
````

````python
class Asset:
  def __init__(self, ticker, price):
    self.ticker = ticker
    self.price = price

class Stock(Asset):
  def __init__(self, ticker, price, high, low, volume):
    super().__init__(ticker, price) #to initialize the parent constructor - in this case asset
    self.high = high
    self.low = low
    self.volume = volume

  def set_price(self, newprice):
    self.price = newprice

class Bond(Asset):
  def __init__(self, ticker, price, maturity, rate):
    super().__init__(ticker, price)
    self.maturity = maturity
    self.rate = rate

  def total_return(self, n):
    return self.price * (1 + self.rate/100) ** n

a = Asset('NFLX', 1000.0)
s = Stock('NVDA', 500.0, 900.0, 400.0, 10)
b = Bond('US_T10', 50.0, 10, 2)
````
